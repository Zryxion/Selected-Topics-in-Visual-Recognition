```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datas = {}
for file in next(os.walk(r'.\loss'))[2]:
    tags = file.split('-')
    if tags[-1][:-4] in datas.keys():
        datas[tags[-1][:-4]][tags[1]] = pd.read_csv(r'.\loss\\' + file)
    else:
        datas[tags[-1][:-4]] = {tags[1]: pd.read_csv(r'.\loss\\' + file)}
```

```python
mapModel = {'Base': 'Mask R-CNN (resnet50)',
            'Base_w_dropout': 'Mask R-CNN (resnet50+drop)',
            'Base_2_dropout': 'Mask R-CNN (resnet101+drop*)'}
```

```python
fig, ax = plt.subplots(2, 3)
for idx, losstype in enumerate(datas.keys()):
    temp = {}
    for model in mapModel.keys():
        temp[mapModel[model]] = datas[losstype][model]['Value']

    ax[int(idx / 3)][int(idx % 3)].set_title(losstype)
    sns.lineplot(data=temp, ax=ax[int(idx / 3)][int(idx % 3)],
                 palette=sns.color_palette()[:3])

fig.set_size_inches(15, 8)
```
